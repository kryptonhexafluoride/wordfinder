```python
# sure i guess we can make this scaleable
n = 4

# parse in word list
words = set()
words_file = open("dictionary.txt", "r")
for word in words_file.readlines():
    word = word.strip();
    words.add(word)
```

```python
# rank frequencies of each letter in each spot
from collections import defaultdict
frequencies = {}
letters = 'abcdefghijklmnopqrstuvwxyz'
for i in range(0, n):
    ith = defaultdict(int)
    for word in words:
        ith[word[i]] += 1
    frequencies[i] = dict(sorted(ith.items()))
print(frequencies)
```

```python
# make a graph
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, n, figsize=(20, 5), sharey=True)
for idx in range(n):
    ax = axs[idx]
    freq = frequencies[idx]
    ax.bar(freq.keys(), freq.values())
    ax.set_title(f'Position {idx+1}')
    ax.set_xlabel('Letter')
    if idx == 0:
        ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
